### Backtracking Algorithm
Testing your algorithm on a sample sudoku is enough, but your code will be manually tested on multiple ones later

```python
def is_empty(sudoku):
    """
    TODO: This function should check whether the sudoku is still empty (contains zero values)
    """
    return np.any(sudoku == 0)
```

```python
def is_placement_valid(grid, num, position):
    """
    TODO: This function determines whether it's valid to place the number 'num' at the given 'position' on the Sudoku grid.
    """
    row, col = position

    # same row
    if num in grid[row]:
        return False

    # same column
    if num in grid[:, col]:
        return False

    # in the 3x3 subgrid
    subgrid_row_start = (row // 3) * 3
    subgrid_col_start = (col // 3) * 3
    if num in grid[subgrid_row_start:subgrid_row_start+3, subgrid_col_start:subgrid_col_start+3]:
        return False

    return True
```

```python
def backtrack(sudoku):
    """
    TODO: implement the backtracking algorithm utilizing is_empty and is_valid functions
    """
    if not is_empty(sudoku):  # no empty cells, puzzle is solved
        return True
    
    for row in range(9):
        for col in range(9):
            if sudoku[row, col] == 0:
                for num in range(1, 10):
                    if is_placement_valid(sudoku, num, (row, col)):
                        # If valid, place the number
                        sudoku[row, col] = num

                        if backtrack(sudoku):
                            return True

                        # If placing num didn't lead to a solution, backtrack
                        sudoku[row, col] = 0

                # no number is valid, return false (backtrack)
                return False

    # no empty cell is found, the puzzle is solved
    return True
```

```python
! python -m pip install -U --user ortools
from ortools.sat.python import cp_model
import pandas as pd
import numpy as np
```

```python
def solve_sudoku(sudoku):
    """
    TODO: Solves the given Sudoku puzzle using ortools and returns the solved grid.
    
    :param sudoku_grid: 2D numpy array representing the Sudoku puzzle
    :return: Solved 2D numpy array or a message indicating failure
    """
    model = cp_model.CpModel()

    # a 9x9 grid 
    grid = [[model.NewIntVar(1, 9, f"cell_{row}_{col}") for col in range(9)] for row in range(9)]

    # constraints for each row and column
    for i in range(9):
        model.AddAllDifferent([grid[i][j] for j in range(9)])  # Row constraint
        model.AddAllDifferent([grid[j][i] for j in range(9)])  # Column constraint

    # 3x3 subgrid
    for box_row in range(3):
        for box_col in range(3):
            subgrid = [
                grid[row][col]
                for row in range(box_row * 3, (box_row + 1) * 3)
                for col in range(box_col * 3, (box_col + 1) * 3)
            ]
            model.AddAllDifferent(subgrid)

    for row in range(9):
        for col in range(9):
            if sudoku[row, col] != 0:
                model.Add(grid[row][col] == sudoku[row, col])

    solver = cp_model.CpSolver()
    status = solver.Solve(model) 

    if status == cp_model.FEASIBLE or status == cp_model.OPTIMAL:
        solved_grid = np.array([[solver.Value(grid[row][col]) for col in range(9)] for row in range(9)])
        return solved_grid
    else:
        return "No solution found"
   
```

### AC-3 Algorithm (20 Points)
First and foremost, implement the AC-3 algorithm to be later used in the problem

```python
def AC3(csp, neighbors):
    """
    csp: 
        dict{
            nodes: list of nodes,
            domains: dictionary {variable: [domain]},
            constraints: dictionary {(node_i, node_j): list of constraints
        },
    neighbors:
        dict {var: [neighbor]}
    """
    
    queue = []
    for node_i in csp['nodes']:
        for node_j in neighbors[node_i]:
            queue.append((node_i, node_j))

    counts = []
    count_for = 0      

    while queue:
        (node_i, node_j) = queue.pop(0)
        revised, csp_new, c = revise(csp, node_i, node_j)
        csp = csp_new

        # If we revise the domain of node_i and remove a value, we need to re-add all neighboring arcs
        if revised:
            if not csp['domains'][node_i]:  # If the domain of node_i is empty after revision, return False
                counts.append(c)
                print("AC3 requires", len(counts), "steps. Average steps:", np.average(counts),". Total operations:", np.sum(counts))
                return False, csp['domains']
                
            for var_k in neighbors[node_i]:
                count_for += 1
                if var_k != node_j:
                    queue.append((var_k, node_i))
        counts.append(max(c, count_for))
    print("AC3 requires", len(counts), "steps. Average steps:", np.average(counts),". Total operations:", np.sum(counts)) 
    
    return True, csp['domains']


def revise(csp, node_i, node_j):
    """
    revises the domain of node_i by removing values that do not satisfy the constraint with node_j.
    
    parameters:
    - csp: The CSP dictionary.
    - node_i, node_j: Variables between which the constraint is checked.
    
    returns:
    - a (bool, csp) with true iff we revise the domain of node_i
    """
    
    revised = False
    count = 0
    for x in csp['domains'][node_i][:]:
        # Check if there eXsts some y in node_j's domain that satisfies the constraint between node_i and node_j
        satisfies_constraint = False
        for y in csp['domains'][node_j]:
            count += 1
            if (x, y) in csp['constraints'][(node_i, node_j)]:
                satisfies_constraint = True
                break  

        # If no supporting y was found, remove x from node_i's domain
        if not satisfies_constraint:
            csp['domains'][node_i].remove(x)
            revised = True
    
    return revised, csp, count
```

### AC-4 Algorithm (20 Points)
Secondly, implement the AC-4 (Arc Consistency 4) algorithm. This algorithm enforces arc consistency by maintaining **support tables** and efficiently processing **a queue of arcs** for consistency checks. To help you implement it, here are the key steps and concepts to focus on:
1. **Support Tables**: For each pair of variable-value assignments, you must maintain a table that tracks the number of supporting values in the neighboring variables' domains. This table will allow you to quickly determine if a value is consistent with its neighbors
2. Start by initializing your support tables for all variable-value pairs. For each arc (X, Y), compute and store the number of values in Y's domain that support each value in X's domain based on the constraint between X and Y.
3. Initialize **a queue of arcs** with all pairs of variables that share a constraint
4. For each arc (X, Y) in the queue, check the support for each value in X's domain. If any value in X loses all supporting values in Y, remove that value from X’s domain.
5. If a value is removed from X’s domain, for every neighboring variable Z (other than Y), you must add the arc (Z, X) back into the queue for further consistency checks. This ensures that removing a value from X doesn’t create inconsistencies for other variables.
6. Whenever a value is removed from a domain, update the support tables for all neighboring variable-value pairs that depend on the removed value. This way, if a value of Z is only supported by the value that was removed from X, the support count for that value of Z should be decremented.
7. The algorithm terminates when the queue is empty, meaning all arcs have been processed and the domains of all variables are arc-consistent. If at any point a domain becomes empty, the problem is unsolvable.

```python
def AC4(csp):
    """
    csp: 
        dict{
            nodes: list of nodes,
            domains: dictionary {variable: [domain]},
            constraints: dictionary {(node_i, node_j): list of constraints
        },
    """
    supported_by = {}
    for x in csp['nodes']:
        for val in csp['domains'][x]:
            supported_by[(x, val)] = []

    queue = []
    counter = {}
    count = 0
    
    for (X, Y) in csp['constraints']:
        for x_val in csp['domains'][X]:
            k = 0
            for y_val in csp['domains'][Y]:
                # Check if (x_val, y_val) satisfies any of the constraints
                count += 1
                if (x_val, y_val) in csp['constraints'][(X, Y)]:
                    k += 1
                    supported_by[(Y, y_val)].append((X, x_val))
            if k == 0:
                csp['domains'][X].remove(x_val)  # Remove unsupported values
                queue.append((X, x_val)) 
                if len(csp['domains'][X]) == 0:
                    print("AC4 required ", count, " steps.")
                    return False, csp['domains']
            else:
                counter[(X, Y, x_val)] = k
    
    count2 = 0
    while queue:
        (X, x_val) = queue.pop(0)
        for (Y, y_val) in supported_by[(X, x_val)]:
            count2 += 1
            counter[(Y, X, y_val)] -= 1
            if counter[(Y, X, y_val)] == 0 and y_val in csp['domains'][Y]:
                csp['domains'][Y].remove(y_val)
                queue.append((Y, y_val))
                if len(csp['domains'][Y]) == 0:
                    print("AC4 required ", max(count, count2), " steps.")
                    return False, csp['domains']

    print("AC4 required ", max(count, count2), " steps.")
    return True, csp['domains']
```

```python
#################################################
############### DO NOT CHANGE ###################
#################################################

graph = {
    "WA": ["NT", "SA"], 
    "NT": ["WA", "SA", "Q"], 
    "SA": ["NT", "WA", "Q", "NSW", "V"], 
    "Q": ["NT", "SA", "NSW"], 
    "NSW": ["Q", "SA", "V"], 
    "V": ["NSW", "SA"]
}

nodes = ["WA", "NT", "SA", "Q", "NSW", "V"]
domains = {}

for node in nodes:
    if node == "WA":
        domains[node] = ["R"]
    else:
        domains[node] = ["R", "G", "B"]

constraint = [(x, y) for x in ["R", "G", "B"] for y in ["R", "G", "B"] if x != y]
constraints = {}

for node in graph.keys():
    for neighbor in graph[node]:
        constraints[(node, neighbor)] = constraint

#################################################
#################################################
#################################################

csp = {
    'nodes': nodes,
    'domains': copy.deepcopy(domains),
    'constraints': constraints,
    'neighbors': graph
}

from time import perf_counter

start_time = perf_counter()
ac3_result, ac3_domains = AC3(csp, graph)
finish_time = perf_counter()
ac3_time = finish_time - start_time

if ac3_result:
    print("AC3 -> Exist consistent solution: ", ac3_domains)
else:
    print("AC3 -> No consistent solution exists")
print(f"solution retrieved in: {ac3_time:.4g} s")


start_time2 = perf_counter()
ac4_result, ac4_domains = AC4(csp)
finish_time2 = perf_counter()
ac4_time = finish_time2 - start_time2

if ac4_result:
    print("AC4 -> Exist consistent solution: ", ac4_domains)
else:
    print("AC4 -> No consistent solution exists")
print(f"solution retrieved in: {ac4_time:.4g} s")
```

### Conclusion
Explain the differences between these two algorithms in terms of time and space complexity. Which one do you prefer to use?

**Your answer**: In AC4, the unsupported values are redundant and can be removed, we keep an additional list of removed values (referring to "queue" in the code) to process them and update the counters. AC4 has 2 other data structures too: "counter" contains arc-value pairs that  has the number of supports each variable gets from each constraint. so when the counter is equal to 0, the value can be removed from its domain. "supported_by" contains all the variable-value pairs that each value of every variable, supports.

for computing AC4 time complexity, if there are maximum k values in the domains, then for each constraint, there are maximum $k^2$ pairs. so if there are n constraints in the problem, the time will be $O(nk^2)$. (The same time for the last while loop.) So AC4's time is $O(nk^2)$ which is less than AC3. This is also evident from the time outputs we observed in the last code block.

But AC4 has a larger space complexity than AC3, which is mainly because of the "supported_by" set ( $O(nk^2)$ ).
So if our problem size is not too large, it's better to use AC4 since it is faster. But for larger problems, using AC4 may lead to memory issues, so we use AC3.

## Backtracking + AC-3 and AC-4 algorithm (20 Points)
Now we are going to utilize the implemented algorithms to speed up the backtracking process, use both of the above algorithms, you must run ac3 and ac4 algorithms seperately.

```python
def is_consistent(var, value, assignment, csp, neighbors):
    for neighbor in neighbors[var]:
        if neighbor in assignment:
            if (value, assignment[neighbor]) not in csp['constraints'][(var, neighbor)]:
                return False
    return True


def select_unassigned_variable(assignment, domains):
    unassigned_vars = [v for v in domains if v not in assignment]
    return min(unassigned_vars, key=lambda var: len(domains[var])) if unassigned_vars else None


def backtracking_search(csp, neighbors, use_ac3):
    """
    Recursive backtracking algorithm with AC3 or AC4 for constraint propagation.
    """
    def backtrack(assignment):
        # If the assignment is complete, return it
        if len(assignment) == len(csp['nodes']):
            return assignment
        
        var = select_unassigned_variable(assignment, csp['domains'])

        for value in csp['domains'][var]:
            if is_consistent(var, value, assignment, csp, neighbors):
                # Assign var = value
                assignment[var] = value
                
                # Perform constraint propagation (either AC3 or AC4)
                if use_ac3:
                    result, new_domains = AC3(csp, neighbors)
                else:
                    result, new_domains = AC4(csp)

                # If arc consistency is maintained, continue
                if result:
                    csp['domains'] = new_domains
                    result = backtrack(assignment)
                    if result:
                        return result
                
                # If failed, remove the assignment and backtrack
                del assignment[var]
        
        return None
    
    # Start backtracking search
    return backtrack({})


csp = {
    'nodes': nodes,
    'domains': domains,
    "constraints": constraints
}

# TODO: Solve using AC3 algorithm

print("\nBacktracking + AC3 algorithm: \n")
start = time()
solution = backtracking_search(csp, graph, True)
s = time() - start

if solution:
    print("Solution found:", solution)
else:
    print("No solution found!")
    
print("Solution retrieved in: %.4g seconds" % s)


# TODO: Solve using AC4 algorithm

print("\nBacktracking + AC4 algorithm:\n")
start = time()
solution = backtracking_search(csp, graph, False)
s = time() - start

if solution:
    print("Solution found:", solution)
else:
    print("No solution found!")
    
print("Solution retrieved in: %.4g seconds" % s)
```
